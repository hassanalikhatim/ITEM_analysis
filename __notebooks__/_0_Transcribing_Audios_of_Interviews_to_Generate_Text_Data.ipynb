```python
import whisper

model = whisper.load_model('base.en')
```

```python
import os

destination_path = ''
source_path = '../Collected Audios/'

list_categories = os.listdir(source_path)
for c, category in enumerate(list_categories):
    print('Working on', category)

    child_path = category + '/'
    list_audios = os.listdir(source_path + child_path)

    for a, audio in enumerate(list_audios):
        file_path = source_path + child_path + audio

        filename = audio.split('.')
        filename = '.'.join(filename[:-1])
        save_path_file = child_path + filename + '.txt'

        if not os.path.isfile(save_path_file):
            result = model.transcribe(file_path)

            with open(save_path_file, 'w') as f:
                f.write(result['text'])
```
